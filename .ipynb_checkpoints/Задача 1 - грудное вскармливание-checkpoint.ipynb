```python
import numpy as np
from scipy.stats import t
import pylab
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
```

Сведения о продолжительности грудного вскармливания в неделях в выборке рожениц:

```python
выборка = [10,22,37,20,31,14,13,19,32,27,15,6,71,17,10,9,17,22,12,20,16,23]
n = len(выборка)
n
```

__Цель__ — оценить среднюю продолжительность вскармливания

__а)__ Рассчитайте 90% доверительный интервал для средней продолжительности, считая
распределение признака нормальным.

```python
альфа = 1 - 0.9
```

Считаем среднюю продолжительность выборки по формуле $\bar X = \frac{\sum_i^n X_i}{n}$

```python
выборочное_среднее = np.sum(выборка)/n
выборочное_среднее
```

Генеральная дисперсия $\sigma^2$ не известна. Тогда заменяем её на выборочную оценку $\hat{\sigma}^2$. 

Считаем несмещенную оценку выборочной дисперсии по формуле $\hat{\sigma}^2 = \frac{\sum_i^n (X_i - \bar{X})^2}{n-1} $

```python
знаменатель1 = 0
for i in range (len(выборка)):
    знаменатель1+= (выборка[i] - выборочное_среднее)**2 
выборочная_дисперсия_2 = знаменатель1/(n - 1)
выборочная_дисперсия_2
```

За счёт этого пропадает одна степень свободы. Тогда получим, что для нашего доверительного интервала нам нужна квантиль $t_{\frac{\alpha}{2}, n-1}$ из распределения Стьюдента с $n-1$ степенями свободы.

```python
квантиль = t.ppf(1 - альфа/2, n-1)
квантиль
```

Наш доверительный интервал выглядит пока так:

$P\left(\bar{X} - t_{\frac{\alpha}{2}, n-1}\frac{\hat{\sigma}}{\sqrt {n}} < \mu < \bar{X} + t_{\frac{\alpha}{2}, n-1}\frac{\hat{\sigma}}{\sqrt {n}}\right) = 1 - \alpha$

То есть наша ошибка: $error = t_{\frac{\alpha}{2}, n-1}\frac{\hat{\sigma}}{\sqrt {n}}$

```python
ошибка = квантиль * (выборочная_дисперсия_2)**(0.5) / (n)**(0.5)
ошибка
```

Теперь просчитаем сам интервал:

```python
выборочное_среднее - ошибка, выборочное_среднее + ошибка
```

Таким образом $[16.031; 26.06] $ - $90\%$ доверительный интервал для $\mu$.

__в)__ Рассчитайте методом Бутстрапа 90% доверительный
интервал для средней продолжительности вскармливания, сгенерировав 1000 перевыборок.

Для метода Бутстрапа нам нужно из уже данной выборки сгенерировать выборки той же длины, рандомно их заполнив. Такие выборки - это перевыборки. От перевыборок нам нужно только выборочное среднее, так что хранить сами перевыборки нет особого смысла. Давайте составим массив из выборочный средних для $1000$ перевыборок:

```python
средние_перевыборок = []
for i in range (1000):
    средние_перевыборок+=[np.sum([random.choice(выборка) for i in range(n)])/n]
```

Так как мы будем брать выборочные квантили из этого набора данных, отсортируем его:

```python
средние_перевыборок.sort()
```

>В общем случае, выборочные
квантили $\hat Q(\frac{\alpha}{2})$ и $\hat Q(1−\frac{\alpha}2)$ для средних в перевыборках образуют бутстраповский
доверительный интервал для среднего в генеральной совокупности с уровнем доверия
$1−\alpha$.

То есть нам нужны квантили $\hat Q(5\%)$ и $\hat Q(95\%)$, то есть:

```python
средние_перевыборок[50],средние_перевыборок[950]
```

Таким образом бутстраповским методом мы получили доверительный интервал $[16.864;25.909]$

Кстати говоря, заметим что доверительный интервал, найденный бутстраповским методом, короче чем тот, что мы нашли в пункте а.
