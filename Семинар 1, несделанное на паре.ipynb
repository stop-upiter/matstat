а) $cov(X,Y) = E(XY) - E(X)E(Y)$

$E(X) = 0\cdot 0.4 + 1\cdot 0.6 = 0.6$

$E(Y) = 0 \cdot 0.25 + 1\cdot 0.5 + 2\cdot 0.25 = 1$

$E(XY) = 0\cdot 0.5 + 1\cdot 0.4 + 2\cdot 0.1 = 0.6$

$cov(X,Y) = 0.6 - 1\cdot 0.6 = 0$

б) Условие независимости: $
P(X=x_i;Y=y_j) = P(X=x_i)P(Y=y_i)
$

Пусть $X = 0$, $Y = 0$, тогда $P(X = 0) = 0.4$, $P(Y = 0) = 0.25$, $P(X = 0)P(Y = 0) = 0.1$, но $P(X=0;Y = 0) = 0.15$ так ещё $P(X = 0; Y = 0) \neq P(X = 0)P(Y = 0)$. Величины зависимые!

Величины независимы, так что запишем условие независимости: $P(X = x_i; Y = y_i) = P(X = x_i)P(Y = y_i)$

Возьмём $X = -1$, $Y = 1$, $P(X = -1) = 0.6 + a$, $P(Y = 1) = 0.25$, $P(X = -1; Y = 1) = 0.2 = P(X = -1)P(Y = 1) = (0.6 + a)\cdot 0.25 = 0.15 + 0.25a$

$0.15 + 0.25a = 0.2$, $0.25a = 0.05$, $a = 0.2$

Возьмём $X = -1$, $Y = 2$, $P(X = -1) = 0.8$, $P(Y = 2) = 0.2 + c$, $P(X = -1; Y = 2) = 0.2 = P(X = -1)P(Y = 2) = 0.8(0.2 + c) = 0.16 + 0.8c$

$0.16 + 0.8c = 0.2$. $0.8c = 0.04$, $c = 0.05$

Полная вероятность $X$: $0.4 + 0.2 + 0.2 + b + 0.05 + 0.05 = 1$, $b = 0.1$

Ответ: $a = 0.2$, $b = 0.1$, $c = 0.05$.
