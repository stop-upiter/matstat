## Семинар 2.
### Свойства оценок.


Есть неизвестный параметр, генеральная совокупность (о которой есть хотя бы представление о распределении), но не знаем реальных значений параметров. Хотим их оценить. Обычно есть выборка, и на ее основе надо оценить параметры.

Есть три свойства оценок: несмещенность, состоятельность, эффективность.

#### Обозначения:
$\theta$ - неизвестный параметр

$\hat{\theta}$ - оценка

$\{\hat{\theta}\}$ - последовательность оценок

#### Свойства:

Пусть есть эмперические величины $X_1, X_2, ..., X_n$.

* Несмещенность: $E(\hat{\theta}) = \theta$

* Состоятельность: $\lim_{n\to\infty}{P(|\hat{\theta} = \theta|>\epsilon)} = 0$
$\lim_{n\to\infty}{P(|\hat{\theta} = \theta|<\epsilon)} = 1$ ($\epsilon>0$)

* Основное (достаточное) условие состоятельности:
    1. Несмещённость, $E(\hat{\theta}) = \theta$, либо асимптотически несмещенность $lim_{n\to\infty}\hat{\theta} = \theta$

    2. $\mathfrak{D}(\hat{\theta}) = 0$ либо $lim_{n\to\infty}\mathfrak{D}(\hat{\theta}) = 0$


* Эффективность: $\mathfrak{D}(\hat{\theta})$ vs $\mathfrak{D}(\tilde{\theta})$

Хотим, чтобы у оценки была как можно меньшая дисперсия. Если, например, $\mathfrak{D}(\hat{\theta}) = \sigma^2$, $\mathfrak{D}(\tilde{\theta}) = \frac{\sigma^2}{n}$,

Relative efficiency (относительная эффективность): $RE = \frac{\mathfrak{D}(\hat{\theta}) [\text{сверху бóльшая дисперсия}]}{\mathfrak{D}(\tilde{\theta})} = \frac{\sigma^2}{\frac{\sigma^2}{n}} = n$

тогда $\tilde{\theta}$ эффективнее, чем $\hat{\theta}$.

* MSE - точность оценки. 
$MSE(\hat{\theta}) = E[(\hat{\theta} - \theta)^2]$

* $\mathfrak{D}(\hat{\theta}) = E(\hat{\theta}^2) - E^2(\hat{\theta})$

* Bias - смещение оценки. 
$Bias(\hat{\theta}) = E(\hat{\theta} - \theta) = [\text{так как }\theta\text{ какая-то числовая характеристика, так ещё константа}] = E(\hat{\theta}) - \theta$

Пусть $z = \hat{\theta} - \theta$, тогда $MSE(\hat{\theta}) = E(z^2)$, $Bias(\hat{\theta}) = E^2(z)$

$\mathfrak{D}(\hat{\theta}) = \mathfrak{D}(\hat{\theta} - \theta) = \mathfrak{D}(z) = E(z^2) - E^2(z)$, 

то есть $E(z^2) = \mathfrak{D}(z) + E^2(z)$, 

а значит $MSE(\hat{\theta}) = \mathfrak{D}(\hat{\theta}) + Bias^2(\hat{\theta})$ $\blacksquare$

а) Дано: $MSE(\hat{\theta}) = MSE(X_1^2) = \mathfrak{D}(X_1^2) + Bias^2(X_1^2)$

$E(X_1^2) = (-1)^2\theta + 0^2(1-2\theta) +1^2\cdot \theta =2\theta =(-1)^4\theta + 0^4(1-2\theta) +1^4\cdot \theta  = E(X^4_1)$

$\mathfrak{D}(X_1^2) = 2\theta - 4\theta^2$

$Bias(X_1^2) = E(\hat{\theta}) - \hat{\theta} = E(X_1^2) - X_1^2 = 2\theta - \theta = \theta$

$MSE(X_1^2) = 2\theta - 4\theta^2 + \theta^2 = 2\theta - 3\theta^2$

б) $\hat{\theta} = c\sum_{i=1}^n X_i^2$, $c=?$ (условие несмещённости: $E(\hat{\theta}) = \theta$)

$E(c\sum_{i=1}^n X_i^2) = c\sum_(i=1)^n E(X_i^2) = cnE(X^2) = [\text{матожидание считали раньше}] = cn2\theta = [\text{по условию несмещенности}] = \theta$

Таким образом $c = \frac{1}{2n}$

в) $MSE(\hat{\theta}) = MSE(\frac{\sum_{i=1}^nX_i^2}{2n}) = [\text{так как оценка несмещенная, то }Bias = 0] = \mathfrak{D}(\frac{\sum_{i=1}^nX_i^2}{2n}) = \frac{\mathfrak{D}(\sum_{i=1}^nX_i^2)}{4n^2} = [\text{так как величины независимые}] = \frac{\sum_{i=1}^n\mathfrak{D}(X_i^2)}{4n^2} =$

$[\text{в силу независимого распределения величин: }\mathfrak{D}(X_i^2) = \mathfrak{D}(X_1^2) = 2\theta - 4\theta^2]= \frac{n(2\theta - 4\theta^2)}{4n^2} = \frac{\theta - 2\theta^2}{2n}$

а) $E(\hat{a}) = E(m) - 3E(X_1 +X_2) = m - 3E(X_1 + X_2) = [\text{так так все величины распределены по одному закону одинаково}] = $

$m - 3E(2X) = m - 6E(X) = m - 6\int_0^1{xf(x)\cdot dx} = m - 6\int_0^1 x(a + 2x(1-a))dx = m - 6\int_0^1(ax + 2x^2 - 2ax^2)dx = $

$m-6(\int_0^1(ax)dx + \int_0^1(2x^2)dx - \int_0^1(2ax^2)dx) = m - 6(\frac{ax^2}{2}|_0^1 + \frac{2x^3}{3}|_0^1 - \frac{2ax^3}{3}|_0^1) =$

$m - 6(\frac{a}{2} + \frac{2}{3} - \frac{2a}{3}) = m - 3a - 4 + 4a = m + a - 4$

Хотим соблюдение условия несмещенности: $E(\hat{a}) = a$:
 
$E(\hat{a}) = a + m -4 = a$ $\to$ $m = 4$.

   0
 \end{cases} $
