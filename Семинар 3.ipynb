Теорема Фишера: если у нас есть набор случайных величин из нормальной генеральной совокупности, где эти $X_1,...,X_n$ независимы, то
1. $\bar{X} \sim N(\mu, \frac{\sigma^2}{n})$
2. оценка $\frac{(n-1)\hat{\sigma}^2}{\sigma^2}$ имеет распределение $\chi^2$ со степенью свободы $n - 1$
3. $\bar{X}$ и $\hat{\sigma}^2$ ($\bar{X}^2$ и $\hat{\sigma}^2$ ) независимы

помним что есть несмещенная оценка выборочной дисперсии
$ \hat{\sigma}^2 = \frac{\sum (X_i-\bar{X})^2}{n-1}
$, 

смещенная оценка выборочной дисперсии $s^2 = \frac{\sum (X_i-\bar{X})^2}{n}$ ($\hat{X}$ и $s^2$ тоже независимы)

нужно быть внимательным при чтении книг, иногда меняют местами обозначения $s^2$ и $\hat{\sigma}^2$

$X \sim N(0,1)$ - стандартное нормальное распределение

$Y \sim N(2,6)$

В конечном итоге хотим все привести к стандному нормальному распределению, чтобы мы могли воспользоваться таблицами

$E(X+2Y) = EX + 2EY = 0 + 2\cdot2 = 4$

$D(X+2Y) = [\text{тк независимы}] = D(X) + 4D(Y) = 1+4*6 = 25$

$X+2Y\sim N(4, 25)$

$P(1<X+2Y<7) = [Z\sim N(0,1)] = P(\frac{1 - E(X+2Y)}{\sqrt{D(X+2Y)}}<Z <\frac{7 - E(X+2Y)}{\sqrt{D(X+2Y)}}) = P(\frac{-3}{5}<Z<\frac{3}{5}) = P(-0.6<Z<0.6)$

Теперь открываем таблицу для $Ф_0$, $P(0<Z<0.6) = 0.2257$, поскольку распеделение симметричное, то $P(-0.6<Z<0.6) = 2P(0<Z<0.6) = 0.4514$ 

Дано: $n=6$

Хотим: $P(\frac{\hat{\sigma}^2}{\sigma^2}>3)$

$\sigma^2$ - дисперсия генеральной совокупности,

$\hat{\sigma}^2 = \frac{\sum (X_i - \bar{X})^2}{n-1}$ - оценка.

$P(\frac{\hat{\sigma}^2}{\sigma^2}>3) = P(\frac{\hat{\sigma}^2(n-1)}{\sigma^2(n-1)}>3(n-1))$ тогда понимаем что величина $\frac{\hat{\sigma}^2(n-1)}{\sigma^2(n-1)}$ имеет распределение $\chi^2$ с степенью свободы $n - 1$

$P(\chi^2>15)$, идём в таблицу хи-квадрата и ищем степень свободы 5 и значение похожее на 15 в таблице, нам нужна альфа. Тогда  $P = 0.01$ 

*таблица хи-квадрата достаточно неточная, ищем что-то близкое или интервал где лежит наше значение.

 $P(\bar{X} > 3.4) = P(Z>\frac{3.4 - E(\bar{X})}{D(\bar{X})}) = P(Z>\frac{3.4-3}{\frac{4}{5}}) = P(Z>0.5)$

 $P(\sum(X_i-\bar{X})^2>43 = P(\chi_{n-1}>\frac{43}{\sigma^2}) = P(\chi_{24}>\frac{43}{16}) = P(\chi_{24}>2.6875) = [\text{по таблице вероятность превысить 10.856 равна 0.99}] > 0.99 \approx 1 $

$X_1, ..., X_n \sim U[-2,2]$ - равномерное распределение

Найти: $lim_{n\to\infty}P(|\sum_{i=1}^nX_i|>3\sqrt{n})$

$\frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt n}} \sim N(0,1)$

по свойствам равномерного распределения:

$E(X_i) = \frac{a+b}{2} = \frac{2-2}{2} = 0$

$D(X_i) = \frac{(b-a)^2}{12} = \frac43$

$\bar{X} = \frac{\sum X_i}{n}$

$P(|\sum_{i=1}^nX_i|>3\sqrt{n}) = P(\frac{|\sum_{i=1}^nX_i|}{n}>\frac{3}{\sqrt{n}}) =P(\frac{ \frac{|\sum_{i=1}^nX_i|}{n} -\mu}{\frac{\sigma}{\sqrt{n}}}>\frac{\frac{3}{\sqrt{n}}-\mu}{\frac{\sigma}{\sqrt{n}}}) = $

$[\text{имеем право занести n под модуль}] =
P(\frac{|\bar{X}| - \mu}{\frac{\sigma}{\sqrt{n}}}>\frac{3 - \mu\sqrt{n}}{\sigma})= [Z\sim N(0,1)] =$

$P(Z>\frac{3 - \mu\sqrt{n}}{\sigma}) = 
P(Z>\frac94) = 0.5 - Ф_0(\frac94) = 0.5 - 0.4878 = 0.0122$
