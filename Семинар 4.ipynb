Оценки бывают точечные и интервальные. Точечная оценка генерального среднего: выборочное среднее. 

#### Построить доверительный интервал для $\mu$, взять доверительную вероятность $1 - \alpha = 90\%$

1 случай. при этом известна дисперсия генеральной совокупности $\sigma^2$ - будет нормальное распределение. Строим на основе $\bar{X}$, $n$, $\sigma^2$. Отталкиваемся от $\bar{X}$.

Ошибка (error) $= Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt n}$

$\bar{X} - Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt n} < \mu <\bar{X} + Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt n}$

$P\left(\bar{X} - Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt n} < \mu <\bar{X} + Z_{\frac{\alpha}{2}}\frac{\sigma}{\sqrt n}\right) = 1 - \alpha$

2 случай. $\sigma^2$ не известна. Тогда заменяем её на выборочную оценку $\hat{\sigma}^2$. За счёт этого пропадает одна степень свободы.

$t_{\frac{\alpha}{2}, n}$ - из распределения Стьюдента с $n-1$ степенями свободы.

$P\left(\bar{X} - t_{\frac{\alpha}{2}, n-1}\frac{\hat{\sigma}}{\sqrt {n}} < \mu < \bar{X} + t_{\frac{\alpha}{2}, n-1}\frac{\hat{\sigma}}{\sqrt {n}}\right) = 1 - \alpha$

###### Обратите внимание, что везде в этих распределения в знаменателе $n$, а не $n-1$!!!

#### Построить доверительный интервал для дисперсии.

Здесь возникает распределение $\chi^2 $, оно не симметрично и очень чувствительно к степеням свободы.

Помним что $\chi^2 = \frac{\hat{\sigma}^2(n-1)}{\sigma^2}$, $\sigma^2$ неизвестно.

$P\left(\frac{\hat{\sigma}^2(n-1)}{\chi_{1-\frac{\alpha}{2}, n-1}^2}<\sigma^2<\frac{\hat{\sigma}^2(n-1)}{\chi_{\frac{\alpha}{2}, n-1}^2}\right)$

В левом знаменателе должно быть бОльшее значение хи-квадрата!

$n = 16$, $\bar{X} = 12.6$

1) $80\%$ доверительный интервал для $\sigma^2$: $1 - \alpha = 0.8$, $\alpha = 0.2$, $\frac{\alpha}{2} = 0.1$, $1 - \frac{\alpha}{2} = 0.9$

Из таблички хи-квадрата: $\chi_{0.1, 15}^2 = 22.307$, $\chi_{0.9, 15}^2 = 8.547$

$P\left(\frac{\hat{\sigma}^2(n-1)}{\chi_{1-\frac{\alpha}{2}, n-1}^2}<\sigma^2<\frac{\hat{\sigma}^2(n-1)}{\chi_{\frac{\alpha}{2}, n-1}^2}\right) = P\left(\frac{6.25\cdot 15}{22.307}<\sigma^2<\frac{6.25\cdot 15}{8.547}\right) = P\left(4.2<\sigma^2<10.96\right)$

$\sigma^2 \in (4.2;10.96)$
