Гипотеза - это некоторое предположение о чём угодно. Гипотеза может быть параметрической и непараметрической.

Пусть все $X \sim N(\mu, \sigma^2)$.

1)

Простая параметрическая гипотеза $H_0$: $\mu = 0$.

Сложная альтернативная гипотеза $H_A$: $\mu \neq 0$.

Как выглядит область принятия гипотезы?

$\alpha$ - уровень значимости. (вероятность ошибки 2 рода)

$P(H_0) = 1 - \alpha$.

$P(H_A) = P(-Z_{\frac{\alpha}{2}}) + P(-Z_{\frac{\alpha}{2}})$

Если $Z < -Z_{\frac{\alpha}{2}}$ либо $Z < Z_{\frac{\alpha}{2}}$ тогда отвергаем $H_0$
