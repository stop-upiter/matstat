Помним про $H_0$ и $H_A$. Обычно нулевая гипотеза, это "что-то чему-то равно", альтернативная - "что-то чему-то не равно" (двусторонняя гипотеза, влияет на то как мы распределяем ошибку $\alpha$, в каждом хвостике по $\frac{\alpha}{2}$) / "что-то меньше/больше значения" (односторонняя, уровень значимости $\alpha$)

$H_0$: $\mu = 450$; $H_A$ : $\mu \neq 450$ (двусторонняя)

$\bar X = \frac{\sum X_i}{n = 12} = 448.25$

считаем несмещённую оценку выборочной дисперсии:
$\hat{\sigma}^2 = \frac1{n - 1}\sum_{i=1}^n(X_i - \bar X)^2 = 14.57$

Берём распределение Стьюдента:
$t = \frac{\bar X - \mu_0}{\sqrt{\frac{\hat{\sigma}^2}{n}}} = \frac{448.25 - 450}{\sqrt{\frac{14.57}{12}}} = -1.588$

Из таблицы находим соответствующее значение критической статистики:
$t_{\frac{\alpha}2, n-1} = t_{0.05, 11} = 1.796$

Статистический критерий:
$|t|>t^{\text{критическое}} \to !H_0$



$1.588 < 1.796 \to H_0$

$H_0$: $\sigma^2 \leq 9$, $H_A$: $\sigma^2 > 9$, $\alpha = 0.01$

Тут у нас хи-квадрат статистика:
$\chi^2 = \frac{(n-1)\hat{\sigma}^2}{\sigma_0^2} = 36$

Если $\chi^2 > \chi^2_{\alpha, n-1} $, то $!H_0$

Из таблицы: $\chi^2_{0.01, 24} = 42.98$

$36 < 42.98 \to \chi^2 \leq \chi^2_{\alpha, n-1} \to H_0$

Приняли нулевую гипотезу.

$n = 25$, $\bar X = 64$, $\hat{\sigma}^2 = 10$, $\alpha = 0.1$

$H_0$: $\mu = 70$, $H_A$: $\mu < 70$

$\mu \geq 70$ это то же самое что $\mu = 70$, просто статистики немного долбанутые и думают что в гипотезах это одно и то же когда это не попадает в альтернативную гипотезу.

$t = \frac{64 - 70}{\frac{10}{\sqrt{25}}} = -3$

$-3 < 1.318 \to H_A$
