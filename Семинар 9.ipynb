Сравнение выборочных средних.

__1)__ Есть связные между собой выборки (matched pairs)
Смысл: взяли группу рабочих на рпедприятии, выбрали, послали на обучение. Вася Петров раньше 1комп/час, теперь 1комп/50минут. Это тольоко Вася так научился или это общая тенденция рабочих на этом курсе? Берём группу из рабочих $n$ и шлём обучаться. Смотрим: какой эффект. 

$H_0$: $\mu_1 = \mu_2$ (нет эффекта) $\to \mu_1 - \mu_2 == 0$

$H_A$: $\mu_1>\mu_2$ (время сборки уменьшилось)

Методология: считаем все разности между до и после для отдельных рабочих, получаем набор разностей:

$D_1, ..., D_n \to \bar D$ - средняя разность, потом нашли выборочную оценку дисперсии для разности $\hat{\sigma}^2_D$

Статистика Стьюдента: 
$t = \frac{\bar D - (\mu_1 - \mu_2)}{\frac{\hat{\sigma}_D}{\sqrt n}} = \frac{\bar D}{\frac{\hat{\sigma}_D}{\sqrt n}}$

$|t|>t_{\text{крит}} \to !H_0$

__2)__ Несвязанные выборки. Взяли из одной и той же ген совокупности выборки, предполагаем $\sigma_1 - \sigma_2 = 0$, но не знаем их значения

$H_0$: $\mu_X = \mu_Y$

Статистика Стьюдента:

$t = \frac{\bar {X} - \bar {Y}}{\sqrt{\frac{n_1s^2_1 + n_2s^2_2}{n_1 + n_2 - 2}}\sqrt{\frac1{n_1} + \frac1{n_2}}} = \frac{\bar {X} - \bar {Y}}{\sqrt{\frac{n_1s^2_1 + n_2s^2_2}{n_1 + n_2 - 2}}\sqrt{\frac{n_1+n_2}{n_1n_2}}}$


Если n большоое, то можно считать что нормальное распределение и  $s^2$ можно считать как $\sigma^2$ (приравниваем выборочную дисперсию дисперсии генеральной совокупности)

__3)__ Известны $\sigma_1^2$, $\sigma_2^2$, $\sigma_1^2 \neq \sigma_2^2$

$t = \frac{\bar X - \bar Y}{\sqrt{\frac{\sigma^2_1}{n_1}+\frac{\sigma_2^2}{n_2}}}$

$H_0$: $\mu_{глуб} \leq \mu_{пов}$

$H_A$: $\mu_{глуб} > \mu_{пов}$

```python
bard = sum(d)/6
bard
```

$\bar D = 0.087$

$\hat{\sigma}^2 = \frac{\sum(D_i - \bar D)^2}{n-1} = 0.0039$

```python
sigma2 = sum([(d[i] - bard)*(d[i] - bard) for i in range(6)])/5
sigma2, sigma2**(0.5)
```

$\hat{\sigma} = 0.063$

```python
t = (bard)/((sigma2/6)**0.5)
t
```
